```python
from openpyxl import load_workbook
from config import get_data_file_path
import pandas as pd

path = get_data_file_path("Opakovanie na TEST.xlsx")
doc = load_workbook(path, data_only=True)
sheet = doc["Kanály"]
data = [[cell.value for cell in row] for row in sheet["A2:F88"]]
frame = pd.DataFrame(data)


frame.sort_values(by=[1], ascending=True, inplace=True)
for i, (index, row) in enumerate(frame.iterrows()):
	frame.loc[index, "anum"] = i + 1

frame.sort_index(inplace=True, axis=0)

frame["adsday"] = frame[5]*24
avgadsday = frame["adsday"].mean() / 60
ads = sheet["G89"].value
adscolor = sheet["G89"].fill.start_color.index
print("Average ads per day:", f'{avgadsday} h')
frame
```
